```python
#converting gender and drink y/n to numerical values
dataset['sex'].replace({'Male':1,'Female':0}, inplace = True)
dataset['DRK_YN'].replace({'Y':1,'N':0}, inplace = True)

correlation = dataset.corr()
sns.heatmap(correlation, xticklabels = dataset.columns, yticklabels = dataset.columns)

```

```python
#columns with clear outliers
weirds = ['waistline','SBP','DBP','BLDS','tot_chole','HDL_chole','LDL_chole','triglyceride','serum_creatinine','SGOT_AST','SGOT_ALT','gamma_GTP']

#looking at correlation data
correlation = dataset.corr()
sns.heatmap(correlation, xticklabels = dataset.columns, yticklabels = dataset.columns)
print(correlation)
```

```python
#removing outliers from the dataset
#chose the IQR method, using a parameter 5


def outlier(feature):
    Q1 = dataset[feature].quantile(0.25)
    Q3 = dataset[feature].quantile(0.75)
    
    IQR = Q3-Q1
    
    lower = Q1 - 5*IQR
    upper = Q3 + 5*IQR
    
    index = dataset[(dataset[feature]<lower) | (dataset[feature]>upper)].index
    print(len(index))
    dataset.drop(index, inplace = True)
    
for column in weirds:
    outlier(column)
    
```

```python
for column in dataset.columns:
    print(column)
    plt.violinplot(dataset[column])
    plt.show()
```

```python
dataset.drop(dataset[(dataset['waistline'] < 50)].index, inplace = True)
```

```python
def new_col(row):
    if row['SMK_stat_type_cd'] == 1:
        if(row['DRK_YN']== 0):
            return 0
        else: 
            return 3
    elif row['SMK_stat_type_cd'] == 2:
        if(row['DRK_YN']== 0):
            return 1
        else:
            return 4
    else:
        if(row['DRK_YN']== 0):
            return 2
        else:
            return 5

dataset["SMK_DRK"] = dataset.apply(new_col,axis = 1)
```

```python
dataset.head()
```

```python
#clearly there are less samples with drinkers and smokers of type 2
#this might skew our result, so i will create a new dataset with similar ratios
X = dataset.drop(columns = ['SMK_stat_type_cd','DRK_YN','SMK_DRK'])
y1 = dataset.DRK_YN
y2 = dataset.SMK_stat_type_cd
y3 = dataset.SMK_DRK

X.describe()
```

```python
from imblearn.under_sampling import RandomUnderSampler

rus1 = RandomUnderSampler(random_state = 1, sampling_strategy = 'majority')

#for drinkers only
X_d,y_d = rus1.fit_resample(X,y1)

#for smokers only
X_s,y_s = rus1.fit_resample(X,y2)

#for both
X_b,y_b = rus1.fit_resample(X,y3)
```

```python
#splitting data into testing and training
from sklearn.model_selection import train_test_split

X_d_train, X_d_test, y_d_train, y_d_test = train_test_split(X_d, y_d, test_size=0.2, random_state=1)
X_s_train, X_s_test, y_s_train, y_s_test = train_test_split(X_s, y_s, test_size=0.2, random_state=1)
X_b_train, X_b_test, y_b_train, y_b_test = train_test_split(X_b, y_b, test_size=0.2, random_state=1)
```

```python
#now i will use a linear regressor to predict
from sklearn.linear_model import LinearRegression


#drunk classifier
drk = LinearRegression()
drk.fit(X_d_train,y_d_train)
y_d_pred = drk.predict(X_d_test)
#rounding it off to make it a classifier
y_d_pred = np.round(y_d_pred)    
```

```python
#smoker classifier
smk = LinearRegression()
smk.fit(X_s_train,y_s_train)
y_s_pred = smk.predict(X_s_test)
#rounding it off to make it a classifier
y_s_pred = np.round(y_s_pred)   
```

```python
#smoker and drinker classifier
bt = LinearRegression()
bt.fit(X_b_train,y_b_train)
y_b_pred = smk.predict(X_b_test)
#rounding it off to make it a classifier
y_b_pred = np.round(y_b_pred)   
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_d_test,y_d_pred))
print(accuracy_score(y_s_test,y_s_pred))
print(accuracy_score(y_b_test,y_b_pred))

#from the accuracy scores its clear that linear regression isnt that great for classification
#and even less so if the number of classes increase
#72.1% for drinkers - 2classes
#57.6% for smokers - 3classes
#9% for both together - 6classes
```

```python
#we will use logistic regression now
from sklearn.linear_model import LogisticRegression

#for drunk
log_drk = LogisticRegression(multi_class='multinomial', solver='saga', max_iter = 4000, penalty = 'l1')
log_drk.fit(X_d_train,y_d_train)
y_d_pred = log_drk.predict(X_d_test)

print(accuracy_score(y_d_test,y_d_pred))

#72%
```

```python
#for smokers
log_smk = LogisticRegression(multi_class='multinomial', solver='saga', max_iter = 4000, penalty = 'l1')
log_smk.fit(X_s_train,y_s_train)
y_s_pred = log_smk.predict(X_s_test)

print(accuracy_score(y_s_test,y_s_pred))

#64.3%
```

```python
#for both
log_b = LogisticRegression(multi_class='multinomial', solver='saga', max_iter = 4000, penalty = 'l1')
log_b.fit(X_b_train,y_b_train)
y_b_pred = log_b.predict(X_b_test)

print(accuracy_score(y_b_test,y_b_pred))

#47.8%

#logistic regression worked better as a classifier, but clearly the data didn't fit this model
#also the execution time was much longer, around 2-12 minutes
```

```python
#now we will try the knn classifier
from sklearn.neighbors import KNeighborsClassifier


#for drunk
knn = KNeighborsClassifier(n_neighbors=13)
  
knn.fit(X_d_train, y_d_train)

y_pred = knn.predict(X_d_test.values)

print(accuracy_score(y_d_test,y_pred))


#69%
```

```python
#for smokers
knn = KNeighborsClassifier(n_neighbors=13)
  
knn.fit(X_s_train, y_s_train)

print(knn.score(X_s_test.values, y_s_test))

#56%
```

```python
#for both
knn = KNeighborsClassifier(n_neighbors=13)
  
knn.fit(X_b_train, y_b_train)

print(knn.score(X_b_test.values, y_b_test))

#41%

#knn took longer to and also had lower accuracies than the logistic regressor
```
